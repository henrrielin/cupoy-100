```python
# 做完特徵工程前的所有準備 (與前範例相同)
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

data_path = 'Day10/data/'
df_train = pd.read_csv(data_path + 'house_train.csv.gz')

train_Y = np.log1p(df_train['SalePrice'])
df = df_train.drop(['Id', 'SalePrice'] , axis=1)
df.head()
```

```python
#只取 int64, float64 兩種數值型欄位, 存於 num_features 中
num_features = []
for dtype, feature in zip(df.dtypes, df.columns):
    if dtype == 'float64' or dtype == 'int64':
        num_features.append(feature)
print(f'{len(num_features)} Numeric Features : {num_features}\n')
```

```python
# 削減文字型欄位, 只剩數值型欄位
df = df[num_features]
df = df.fillna(-1)
MMEncoder = MinMaxScaler()
train_num = train_Y.shape[0]
df.head()
```

# 作業1
* 試著限制 '1樓地板面積(平方英尺)' (1stFlrSF) 欄位的上下限, 看看能否再進一步提高分數?

```python
# 顯示 1stFlrSF 與目標值的散佈圖
import seaborn as sns
import matplotlib.pyplot as plt
sns.regplot(x = df['1stFlrSF'][:train_num], y=train_Y)
plt.show()

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 調整離群值
"""
Your Code Here
"""
X = df.copy()
X['1stFlrSF'] = X['1stFlrSF'].clip(500,2000)

# 做線性迴歸, 觀察分數
X = MMEncoder.fit_transform(X)
estimator = LinearRegression()
cross_val_score(estimator, X, train_Y, cv=5).mean()
```

# 作業2
* 續前題, 去除離群值有兩類方式 :  捨棄離群值(刪除離群的資料) 以及調整離群值,  
請試著用同樣的上下限, 改為 '捨棄離群值' 的方法, 看看結果會變好還是變差? 並試著解釋原因。

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 捨棄離群值
"""
Your Code Here
"""
X = df.copy()
Y = train_Y[(X['1stFlrSF'] > 500) & (X['1stFlrSF'] < 2000)]
X = X[(X['1stFlrSF'] > 500) & (X['1stFlrSF'] < 2000)]


# 做線性迴歸, 觀察分數
X = MMEncoder.fit_transform(X)
estimator = LinearRegression()
cross_val_score(estimator, X, Y, cv=5).mean()
```

### 解釋原因
捨棄離群值在回歸預測上會有比較好的效果，原因是因為離群值是資料裡面少數的異常值，相比直接調整離群值到限定範圍會更好。
調整離群值的隱憂是將實際資料做異動會造成資料與現實不符，如果真實世界異常值是真實存在的，某方面是在資料上作弊（將真實資料修正為假資料好應用預測）
